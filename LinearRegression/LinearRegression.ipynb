```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
employee_df = pd.read_csv("Employee.csv")
employee_df.head()
```

```python
x = employee_df.age.values.reshape(-1,1)
y = employee_df.salary.values
employee_lr = LinearRegression()
employee_lr.fit(x, y)
```

```python
print("Intercept - c:", employee_lr.intercept_)
print("Coeff - m:", employee_lr.coef_)
```

```python
plt.plot(x, employee_lr.predict(x), color='blue', linewidth=3)
```

```python
predict = employee_lr.predict([[35]])
predict
```
